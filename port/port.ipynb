# 端口状态

```python
#coding:UTF-8
import os
import socket
def IsOpen(ip,port):
    s = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
    try:
        s.connect((ip,int(port)))
        s.shutdown(2)
        print '%d is open' % port
        return True
    except:
        print '%d is close' % port
        return False
if __name__ == '__main__':
    n=int(raw_input(u"输入端口号："))
    IsOpen('127.0.0.1',n)
```

# 端口被哪个程序占用

### 所有端口被占用情况

```python
#os.system('netstat -ano') 执行cmd命令，但不会返回结果
f=os.popen('netstat -ano') #将执行cmd命令的返回结果存入file类型的变量
#print f.read()
r=f.read()
```

```python
#去除返回之内空格等字符
all_port=r.split(' ')
all_port=all_port[all_port.index('TCP'):]
all_port=[i for i in all_port if len(i)!=0]
```

### 某个端口被占用PID

```python
n=raw_input(u'请输入端口号：')
```

```python
ff=os.popen('netstat -aon|findstr '+str(n))
rr=ff.read()
```

```python
n_port=rr.split(' ')
n_port=[i for i in n_port if len(i)!=0]
```

```python
#输出端口对应的pid
for i in range(1,len(n_port),5):
    if n_port[i].endswith(n):
        pid = n_port[i+3]
        print pid
        break
```

### 被占用端口对应的程序

```python
import psutil
p = psutil.Process(int(pid))
print('Process name : %s' % p.name())
```

### 杀死该进程

```python
p.kill()
```
