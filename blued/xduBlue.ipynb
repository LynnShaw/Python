```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import plotly.plotly as py
import plotly.graph_objs as go
import cufflinks as cf

cf.set_config_file(offline=False, world_readable=True, theme='ggplot')
```

```python
dataCsv=pd.read_csv('xduBlue.csv',sep='\t', encoding='utf-8')
dataCsv[dataCsv.distance<2]#取distance 小于3的行,统计01分布
ageSeries = dataCsv[dataCsv.distance<0.5]['age'].value_counts()
ageDF1 = pd.DataFrame(ageSeries[ageSeries.index<30])
ageDF1 = ageDF1.drop([17,18,19])
ageDF1.loc['>=30']=[sum(ageSeries[ageSeries.index>=30])]
ageDF1.loc['<=20']=[sum(ageSeries[ageSeries.index<=20])]
ageDF1
ageSeries = dataCsv[dataCsv.distance<1]['age'].value_counts()
ageDF2 = pd.DataFrame(ageSeries[ageSeries.index<30])
ageDF2 = ageDF2.drop([17,18,19])
ageDF2.loc['>=30']=[sum(ageSeries[ageSeries.index>=30])]
ageDF2.loc['<=20']=[sum(ageSeries[ageSeries.index<=20])]
ageDF2
ageSeries = dataCsv[dataCsv.distance<2]['age'].value_counts()
ageDF3 = pd.DataFrame(ageSeries[ageSeries.index<30])
ageDF3 = ageDF3.drop([17,18,19])
ageDF3.loc['>=30']=[sum(ageSeries[ageSeries.index>=30])]
ageDF3.loc['<=20']=[sum(ageSeries[ageSeries.index<=20])]
ageDF3
ageSeries = dataCsv[dataCsv.distance<3]['age'].value_counts()
ageDF4 = pd.DataFrame(ageSeries[ageSeries.index<30])
ageDF4 = ageDF4.drop([17,18,19])
ageDF4.loc['>=30']=[sum(ageSeries[ageSeries.index>=30])]
ageDF4.loc['<=20']=[sum(ageSeries[ageSeries.index<=20])]
ageDF4
```
