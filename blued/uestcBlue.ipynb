```python
dataCsv=pd.read_csv('uestcBlue.csv',sep='\t', encoding='utf-8')
```

```python
figureDF=dataCsv.head(666)[['role','height','weight']]
figureDF=figureDF[figureDF.height<=195]
figureDF=figureDF[figureDF.weight<=140]
figureDF=figureDF[figureDF.height>=155]
figureDF=figureDF[figureDF.weight>=40]
```

```python
figureDF0=figureDF[figureDF.role==0.0]
figureDF1=figureDF[figureDF.role==1.0]
figureDF5=figureDF[figureDF.role==0.5]
figureDF_1=figureDF[figureDF.role==-1]
```

```python
import plotly.plotly as py
import plotly.graph_objs as go
import math
import numpy as np

trace0 = go.Scatter(
    y = list(figureDF0['height']),
    x = list(figureDF0['weight']),
    name = '0',
    mode = 'markers',
    marker = dict(
        size = 8,
        color = 'rgba(152, 0, 0, .8)',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)

trace1 = go.Scatter(
    y = list(figureDF1['height']),
    x = list(figureDF1['weight']),
    name = '1',
    mode = 'markers',
    marker = dict(
        size = 8,
        color = 'rgba(255, 182, 193, .9)',
        line = dict(
            width = 2,
        )
    )
)
trace2 = go.Scatter(
    y = list(figureDF5['height']),
    x = list(figureDF5['weight']),
    name = '0.5',
    mode = 'markers',
    marker = dict(
        size = 8,
        color = 'rgba(64, 116, 52, .8)',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)

trace3 = go.Scatter(
    y = list(figureDF_1['height']),
    x = list(figureDF_1['weight']),
    name = '~',
    mode = 'markers',
    marker = dict(
        size = 8,
        color = 'rgba(62, 18, 202, .5)',
        line = dict(
            width = 2,
        )
    )
)
x=range(40,100,10)
x_rev = x[::-1]
y1_upper=[round(math.sqrt(i/18.5)*100) for i in x]
y1_lower=[round(math.sqrt(i/25.0)*100) for i in x]
y1_lower = y1_lower[::-1]
trace4 = go.Scatter(
    x=x+x_rev,
    y=y1_upper+y1_lower,
    fill='tozerox',
    fillcolor='rgba(0,100,80,0.2)',
    line=go.Line(color='transparent'),
    showlegend=False,
    name='Fair',
)

data = [trace0, trace1,trace2,trace3,trace4]
layout = go.Layout(
    title='èº«é«˜ä½“é‡åˆ†å¸ƒè¡¨',
    paper_bgcolor='rgb(255,255,255)',
    plot_bgcolor='rgb(229,229,229)',
    xaxis=go.XAxis(
        title='ä½“é‡ï¼ˆkgï¼‰',
        gridcolor='rgb(255,255,255)',
        range=[40,100],
        showgrid=True,
        showline=False,
        showticklabels=True,
        tickcolor='rgb(127,127,127)',
        ticks='outside',
        zeroline=False
    ),
    yaxis=go.YAxis(
        title='èº«é«˜ï¼ˆcmï¼‰',
        range=[150,200],
        gridcolor='rgb(255,255,255)',
        showgrid=True,
        showline=False,
        showticklabels=True,
        tickcolor='rgb(127,127,127)',
        ticks='outside',
        zeroline=False
    ),
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='styled-scatter',)
```
